```python
from dataclasses import dataclass

@dataclass
class Packet:
    data: str
    parity_error: bool
    computed_crc: str
    crc_error: bool
```

```python
import re

value_lines = []
with open("../MonitorCaptures/capture1_08272021_08_49_pm.txt", "r") as f:
    for line in f:
        rslt = re.search(r"\* [0-9]\.[0-9]{3}: ([0-9A-F]+)(-PE:)? CRC=([0-9A-F]{2})( CRC error)?", line)
        if rslt:
            packet = Packet(data=rslt.groups()[0], parity_error=(rslt.groups()[1] != None), computed_crc=rslt.groups()[2], crc_error=(rslt.groups()[3] != None))
            value_lines.append(packet)
```

```python
# Proportion of packets with parity errors
len([x for x in value_lines if x.parity_error]) / len(value_lines)
```

```python
packets = value_lines
non_pe_packets = [x for x in value_lines if not x.parity_error]
pe_packets = [x for x in value_lines if x.parity_error]
```
